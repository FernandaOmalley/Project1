```python
# Imports
import plotly.express as px
import matplotlib.pyplot as plt
import panel as pn
import pandas as pd
import os
import seaborn as sns
from pathlib import Path
from dotenv import load_dotenv

%matplotlib inline
```

```python
# Set up Panel Plotly extension
pn.extension('plotly')
```

```python
# Import hvplot.pandas after pn.extension
import hvplot.pandas
```

```python
# Reading data

# Reading House Price Index (HPI) dataset
hpi_path = Path("Data/hpi_by_housing_type_2005_2021.csv")
df_hpi= pd.read_csv(hpi_path, index_col="Date",infer_datetime_format=True, parse_dates=True)
df_hpi=df_hpi[["Composite_HPI_SA", "Composite_Benchmark_SA"]].dropna().sort_index()

# Reading Consumer Price Index (CPI) dataset
cpi_path = Path("Data/cpi_ontario.csv")
df_cpi= pd.read_csv(cpi_path, index_col="REF_DATE",infer_datetime_format=True, parse_dates=True).reset_index()
df_cpi = df_cpi[["REF_DATE","Products and product groups","VALUE"]]
df_cpi = df_cpi.loc[df_cpi["Products and product groups"]=="All-items"].drop(columns="Products and product groups")
df_cpi.columns=["Date","Consumer Price Index (CPI)"]
df_cpi = df_cpi.set_index("Date").sort_index().dropna()

# Reading primary rate dataset
prate_path = Path("Data/primary_rate_canada.csv")
df_prate= pd.read_csv(prate_path, index_col="Date",infer_datetime_format=True, parse_dates=True).dropna().drop(columns="Bank of Canada Overnight Rate").sort_index()

# Reading unemployment rate dataset
unemployment_path = Path("Data/unemployment_rate_age15to64_ontario.csv")
df_unemployment= pd.read_csv(unemployment_path, index_col="Date",infer_datetime_format=True, parse_dates=True)
df_unemployment = df_unemployment.loc[df_unemployment["Labour force characteristics"]=="Unemployment rate"]
df_unemployment = df_unemployment[["VALUE"]]
df_unemployment.columns=["Unemployment Rate"]
df_unemployment = df_unemployment.sort_index().dropna()

# Reading household income dataset
income_path = Path("Data/household_income_by_family_type_2000_2018.csv")
df_income= pd.read_csv(income_path, index_col="YEAR")
df_income = df_income.loc[df_income["Family characteristics"]=="Median total income, all families"]
df_income = df_income[["VALUE"]]
df_income.columns=["Household Income"]
df_income = df_income.sort_index().dropna()
```

```python
# Calculate yearly values for each variables
df_cpi_yr = df_cpi.groupby([df_cpi.index.year]).mean().round(2)
df_hpi_yr = df_hpi.groupby([df_hpi.index.year]).mean().round(2)
df_prate_yr = df_prate.groupby([df_prate.index.year]).mean().round(2)
df_unemployment_yr = df_unemployment.groupby([df_unemployment.index.year]).mean().round(2)
df_income_yr = df_income
```

```python
# Combine all datasets
df_combined = pd.concat([df_cpi_yr,df_income_yr,df_prate_yr,df_unemployment_yr,df_hpi_yr], axis=1).dropna()
df_combined.columns = ["Consumer Price Index (CPI)", "Household Income", "Prime Rate", "Unemployment Rate", "House Price Index (HPI)", "Benchmark Housing Value"]
df_combined.index.set_names("Year", inplace=True)
df_combined
```

```python
# Calculate yearly changes
df_combined_yr_changes = df_combined.pct_change().dropna()
df_combined_yr_changes.index.set_names("Year", inplace=True)
df_combined_yr_changes
```

```python
# Plot the trends for all variables
df_combined_yr_changes.hvplot(xlabel="Year", ylabel="Change Percentage",rot=45)
```

```python
# Calculate and plot the correlation for all variables
correlation_combined = df_combined_yr_changes.corr()
sns.heatmap(correlation_combined, vmin=-1, vmax=1, annot=True)
```

```python
# Export yearly values for all macro variables as CSV files
df_combined.to_csv("Data/yearly_macro_variables_summary.csv")
df_combined_yr_changes.to_csv("Data/yearly_macro_variables_changing_pct_summary.csv")
```
